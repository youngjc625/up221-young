```python
# get values in first 5 rows for the B17020_calc_numChildPovE column 
# Child Population (under 18 years) whose income in the past 12 months is below poverty level
poverty['B17020_calc_numChildPovE'].head()
```

```python
# get values in first 5 rows for the B17020_calc_pctChildPovE column 
# Percent of Children (under 18 years) whose income in the past 12 months is below poverty level
poverty['B17020_calc_pctChildPovE'].head()
```

```python
# show a dataframe with a subset of columns
poverty[['GEOID','NAME','B17020_calc_numChildPovE','B17020_calc_pctChildPovE','geometry']]
```

```python
# list of desired column names
desired_columns = ['GEOID','NAME','B17020_calc_numChildPovE','B17020_calc_pctChildPovE','geometry']

# subset based on desired columns
poverty[desired_columns].sample(5)
```

```python
# DECLARE as new variable to preserve new dataframe and make a copy
poverty_trimmed = poverty[desired_columns].copy()
poverty_trimmed
```

```python
# renaming columns for more understandable data table
poverty_trimmed.columns = ['FIPS Code','Census Tract','# Children in Poverty','% Children in Poverty','geometry']
poverty_trimmed
```

```python
# shows any repeats of the # of children in poverty, 
# which in this case is mostly useful for finding out how many census tracts have 0 children in poverty
poverty_trimmed['# Children in Poverty'].value_counts()
```

```python
# filtering out tracts with 0 child poverty
filtered_poverty_trimmed = poverty_trimmed[(poverty_trimmed != 0).all(axis=1)]
filtered_poverty_trimmed
```

```python
# visualizing data
filtered_poverty_trimmed.plot()
```
